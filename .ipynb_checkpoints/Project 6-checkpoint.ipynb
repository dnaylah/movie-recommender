# clean and refine data

```python
#get website information
soup = BeautifulSoup(urllib2.urlopen('http://www.imdb.com/chart/top'))
```

```python
#grab table that holds all the movie info
table = soup.find_all('tbody', class_='lister-list')
```

```python
#make empty list to hold all the info
movie_ids = []
```

```python
for id1, row in enumerate(table):
    #finding all poster images
    for id2, hello in enumerate(row.find_all('td', class_='posterColumn')):
        #print img['src']
        for id3, a in enumerate(hello.find_all('a')):
            movie_ids.append(a['href'].split("/")[2])
        
```

```python
reviews4 = []
for item in DFimdb.imdbID:
    string = ''
    s = BeautifulSoup(urllib2.urlopen('http://www.imdb.com/title/'+item+'/reviews?ref_=tt_urv').read(), "lxml")
    for ele in s.find_all('div', id= "tn15content"):
        for x in ele.find_all('p', text=True, recursive=False):
            string += x.text
 
    reviews4.append(string)
    
```

```python
s1 = pd.DataFrame(reviews4)
```

```python
DFimdb['Reviews'] = [x.replace("\n","")
                     .replace("Add another review","").replace("*","")
                     .replace("\r\n","").replace("\r","") for x in reviews4]
```

```python
DFimdb.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer(max_features = 50,
                             ngram_range=(1, 1),
                             stop_words='english',
                             binary=True)

X_reviews = pd.DataFrame(v.fit_transform(DFimdb['Reviews']).todense(), columns=v.get_feature_names())
X_reviews.head()
```

```python
v = CountVectorizer(max_features = 100,
                             ngram_range=(1, 1),
                             stop_words='english',
                             binary=True,
                             token_pattern=u'(?u)\\b\\w\\w+\\b')

X_Country = v.fit_transform(DFimdb.Country)
Country_df = pd.DataFrame(X_Country.todense(), columns=v.get_feature_names())
```

```python
Rated_df= pd.get_dummies(DFimdb.Rated)
```

```python
finalDF = pd.concat([DFimdb, X_reviews, Country_df, Rated_df], axis=1, join_axes=[DFimdb.index])
```

```python
finalDF.to_csv('project_imdb.csv', sep='\t', encoding='utf-8')
```

```python
#create features and target
y = finalDF['imdbRating']
X = finalDF.iloc[:,196:]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)


```

```python
dt = DecisionTreeRegressor()

scrs = cross_val_score(dt, X, y, cv=3)
print "cross Val Score: ", np.mean(scrs)

dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)
plt.scatter(y_pred, y_test)
plt.show()
print "R2 Score: ", dt.score(X_test, y_test)
```

```python
GBR = GradientBoostingRegressor(random_state=1)
scrs = cross_val_score(GBR, X, y, cv=3)
print "cross Val Score: ", np.mean(scrs)

GBR.fit(X_train, y_train)

y_pred = GBR.predict(X_test)
plt.scatter(y_pred, y_test)
plt.show()
print "R2 Score: ", GBR.score(X_test, y_test)
```

```python
ss = StandardScaler()

X = ss.fit_transform(DFimdb._get_numeric_data().drop("imdbRating",axis=1))
y = DFimdb['imdbRating']
```

```python
new_df = pd.DataFrame(X, columns = DFimdb._get_numeric_data().drop("imdbRating", axis=1).columns)
new_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
dt =  DecisionTreeRegressor(random_state=1)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
print "Decision tree score: ",dt.score(X_test, y_test)
```

```python
bdt = BaggingRegressor(dt, random_state=1)
bdt.fit(X_train,y_train)
print "Bagging dt score: ", bdt.score(X_test,y_test)
```

```python
rfr = RandomForestRegressor(random_state=1)
rfr.fit(X_train,y_train)
print "random forest score: ", rfr.score(X_test, y_test)
```

```python
abr = AdaBoostRegressor(random_state=1)
abr.fit(X_train,y_train)
print "ada boost score: ", abr.score(X_test,y_test)
```

```python
gbr = GradientBoostingRegressor(random_state=1)
gbr.fit(X_train, y_train)
print "gradient boost score: ", gbr.score(X_test,y_test)
```

```python
#find number of best features
highest_score = 0
highest = 5
for i in range(5,180):
    gbr = GradientBoostingRegressor(random_state=1)
    #iterate through k's
    selector = SelectKBest(f_classif, k=i)
    selected_data = selector.fit_transform(X, y)
    temp = pd.DataFrame(selected_data)
    X_train, X_test, y_train, y_test = train_test_split(temp, y, test_size=0.33, random_state=42)
    
    gbr.fit(X_train, y_train)
    
    score = gbr.score(X_test, y_test)
    
    if score > highest_score:
        highest_score = score
        highest = i
print "Kbest features %d with max score %f " %(highest, highest_score)
```

```python
#train /test off that number
selector = SelectKBest(f_classif, k=52)

selected_data = selector.fit_transform(X, y)
selectDF = pd.DataFrame(selected_data)

X_train, X_test, y_train, y_test = train_test_split(selectDF, y, test_size=0.33, random_state=42)
```

```python
gbr = GradientBoostingRegressor(random_state=1)
gbr.fit(X_train, y_train)
y_pred = gbr.predict(X_test)
print "gbr score: ", gbr.score(X_test, y_test)

plt.figure(figsize=(10,6))
sns.regplot(y_pred,y_test)
plt.title("actuals vs predictions for imdb ratings")
plt.xlabel("predicted")
plt.ylabel("actual")
plt.show()
```
